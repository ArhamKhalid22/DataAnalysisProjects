```python

full_df = pd.DataFrame()
path = '/content/additional_data.zip'

with zipfile.ZipFile(path, 'r') as zip_ref:
    for file in file_csv:
        # Extract the filename from the path within the zip
        filename = file.split('/')[-1]

        # Read the CSV file from the zip archive, specifying the encoding
        with zip_ref.open(file) as f:
            current_df = pd.read_csv(f, encoding='latin1') # or 'utf-16', 'cp1252', etc.
                                                           # Try different encodings if 'latin1' fails

        full_df = pd.concat([full_df, current_df], ignore_index=True)
```

```python
import zipfile
import pandas as pd
import io

# Specify the path to the zip file and the JSON file you want to read
zip_file_path = '/content/additional_data.zip'
json_file_name = 'additional_data/US_category_id.json'  # Replace with the desired JSON file

# Open the zip file
with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    # Read the contents of the JSON file into a string
    with zip_ref.open(json_file_name) as json_file:
        json_data = json_file.read().decode('utf-8')  # Assuming the JSON file is UTF-8 encoded

# Use pandas to read the JSON string into a dataframe
df = pd.read_json(io.StringIO(json_data)) # Wrap the json_data string with io.StringIO to make it work.

# Now you can work with the dataframe 'df'
print(df.head())
```

```python
def punctuation_count(text):
  return len([char for char in text if char in string.punctuation])
```

```python
sample = full_df[0:10000]
```

```python
sample['count_punc'] = sample['title'].apply(punctuation_count)
```
